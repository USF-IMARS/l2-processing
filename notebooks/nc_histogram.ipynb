```python
# netcdf viewing

from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt

# my_example_nc_file = './MODA_OC_py_data/A2007143182500.L2_LAC_OC.x.nc'  # big file
my_example_nc_file = './MODA_OC_py_data/A2007143183000.L2_LAC_OC.x.nc'  # small file
# my_example_nc_file = './MODA_OC_py_data/A2007146203500.L2_LAC_OC.x.nc'  # tiny file

with Dataset(my_example_nc_file, mode='r') as filehandle:    
    chlor = filehandle.groups['geophysical_data'].variables['chlor_a'][:]  # TODO: can we slice this?

# TODO: plot histogram & trim boundaries
#       eg: https://numpy.org/doc/stable/reference/generated/numpy.histogram.html
# np.histogram(chlor)
# display(chlor)

# from: https://numpy.org/doc/stable/reference/generated/numpy.histogram.html
_ = plt.hist(chlor.ravel(), bins='auto')  # .ravel() collapses the 2d array into 1d
plt.title("Histogram with 'auto' bins")
# Text(0.5, 1.0, "Histogram with 'auto' bins")


#     fig = plt.gcf()
#     fig.set_size_inches(30,30)
#     # fig.savefig('test2png.png', dpi=100)
fig = plt.gcf()
fig.set_size_inches(20,5)
# plt.xlim([0,2])
plt.show()
```
